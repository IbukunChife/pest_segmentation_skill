```python
import os
import cv2
import h5py
import random
import numpy as np
from skimage import img_as_float32
from sklearn.base import BaseEstimator
from sklearn.base import TransformerMixin
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
class SegmentationProcessing(BaseEstimator,TransformerMixin):
    
    def __init__(self, path_in, path_out, attributes, drawing, top_left_pt, bottom_right_pt):
        self.path_in = path_in
        self.path_out = path_out
        self.attributes = attributes
        self.dados_X = []
        self.dados_y = []
        self.x_init = None
        self.y_init = None
        self.drawing = drawing
        self.top_left_pt = top_left_pt
        self.bottom_right_pt = bottom_right_pt
        self.image_original = None
        self.image_copy = None
        self.contador = 0
        self.rect = None
        self.attr_moment = ""
    
    def fit(seft,*_):
        return self
    
    def draw_rectangle(self, event, x, y, flags, params):
        # Detecting mouse button down event
        if event == cv2.EVENT_LBUTTONDOWN:
            self.drawing = True
            self.x_init, self.y_init = x, y

        # Detecting mouse movement
        elif event == cv2.EVENT_MOUSEMOVE:
            if self.drawing:
                top_left_pt, bottom_right_pt = (self.x_init, self.y_init), (x, y)
                self.image_copy[self.y_init:y, self.x_init:x] = 255 - self.image_original[self.y_init:y,self.x_init:x]
                cv2.rectangle(self.image_copy, top_left_pt, bottom_right_pt, (0, 255, 0), 2)

        # Detecting mouse button up event
        elif event == cv2.EVENT_LBUTTONUP:
            self.drawing = False
            top_left_pt, bottom_right_pt = (self.x_init,self.y_init), (x, y)
            self.image_copy[self.y_init:y, self.x_init:x] = 255 - self.image_copy[self.y_init:y, self.x_init:x]
            cv2.rectangle(self.image_copy, top_left_pt, bottom_right_pt, (0, 255, 0), 2)
            rect_final = (self.x_init,self.y_init, x-self.x_init, y-self.y_init)
            self.rect = rect_final
        
        # Run Grabcut on the region of interest
        self.run_grabcut()
        
    def run_grabcut(self,*_):
        # Initialize the mask
        mask = np.zeros(self.image_original.shape[:2], np.uint8)

        # Extract the rectangle and set the region of
        # interest in the above mask
        x, y, w, h = self.rect
        mask[y:y+h, x:x+w] = 1

        # Initialize background and foreground models
        bgdModel = np.zeros((1, 65), np.float64)
        fgdModel = np.zeros((1, 65), np.float64)

        # Run Grabcut algorithm
        cv2.grabCut(self.image_original, mask, self.rect, bgdModel,
                    fgdModel, 5, cv2.GC_INIT_WITH_RECT)

        # Extract new mask
        mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')

        # Apply the above mask to the image
        self.image_original = self.image_original*mask2[:, :, np.newaxis]
        
        # Save the image
        cv2.imwrite("{}/{}/{}.jpg".format(self.path_out,self.attr_moment,self.contador),self.image_original)
        self.contador +=1 
        
#         cv2.destroyAllWindows()

        # Display the image
        # cv2.imshow('{}'.format(self.contador), img_orig)
    
    
    def get_files(self, attribute):
        files = sorted([os.path.join(self.path_in, attribute, file)
                        for file in os.listdir(self.path_in + "/"+attribute)
                        if file.endswith('.jpg')])
        random.shuffle(files)
        return files    
    
    def load_image(self,item):
        
        self.image_original = cv2.imread(item)
        self.image_original = cv2.resize(self.image_original, None, fx=0.5, fy=0.5,interpolation=cv2.INTER_CUBIC)
        self.image_copy = self.image_original.copy()
        
        cv2.namedWindow('Input')
        cv2.setMouseCallback('Input', self.draw_rectangle)
        proximo = 1
        while proximo:
            cv2.imshow('Input', self.image_copy)
            c = cv2.waitKey(1000)
            if c == 27:
                break
            proximo = 0
#             break
#         cv2.destroyAllWindows()
#         return 
#         cv2.destroyAllWindows()

    
        
    def make_sets(self, *_):
        for attribute in self.attributes:
            self.attr_moment = attribute
            
            dados = self.get_files(attribute)
            
            for item in dados:
                image = self.load_image(item)
            self.contador = 0
            
        
    def transform(self,*_):
        self.make_sets()
        
```

```python
path_in = "Imagens/insectos_in"
path_out = "Imagens/insectos_out"
attributes = ["bliPuch","flavi","larg"]
drawing = False
top_left_pt, bottom_right_pt = (-1, -1), (-1, -1)
crop = SegmentationProcessing (path_in, path_out, attributes,drawing, top_left_pt, bottom_right_pt)
```
